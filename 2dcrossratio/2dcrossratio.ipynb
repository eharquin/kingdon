```python
from kingdon import Algebra
import math
import time

alg = Algebra(2, 0, 1)
locals().update(alg.blades)
```

```python
# Construct points and lines from coefficients
point = lambda x, y: (e0 + x*e1 + y*e2).dual()
area = lambda a,b,c: a & b & c
#crossRatio2d = lambda a,b,c,d,e: ( area(e,d,c) | area(e,b,a) ) / ( area(e,a,c) | area(e,b,d) )
crossRatio2d = lambda a,b,c,d,e: ( area(a,b,d) | area(a,c,e) ) / ( area(a,c,d) | area(a,b,e) )
```

```python
A = point(-3, -1)
B = point(-3,  1)
C = point( -1,  1)
D = point( -1, -1)
E = point( -2,  0.5)
```

```python
k = crossRatio2d(A,B,C,D,E)
```

```python
alg.graph(
    "Drag the points A,B,C,D,E",                        # render title
    0x224488, A, "A", B, "B", C, "C", D, "D", E, "E",         # render points C1    
    lineWidth=3, grid=True, labels=True, animate=True,
)
```
