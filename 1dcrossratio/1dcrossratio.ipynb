```python
from kingdon import Algebra
import math
import time

alg = Algebra(2, 0, 1)
locals().update(alg.blades)

point = lambda x, y: (e0 + x*e1 + y*e2).dual()
line = lambda a, b, c: (c*e0 + a*e1 + b*e2)
```

```python
crossRatio = lambda a,b,c,d: ( (a & c) * ( b & d) ) / ( (b & c) * ( a & d) ) 
```

# Point Cross-Ratio

```python
O = e12
V = -e02

t1 = 0.5
t2 = 0.8
t3 = 1
t4 = 2

P1 = O + t1*V
P2 = O + t2*V
P3 = O + t3*V
P4 = O + t4*V



k = lambda: f"cross-ratio : {crossRatio(P1,P2,P3,P4):.3f}" 
```

```python
k()
```

```python
alg.graph(
    "Drag the points A,B,C,D",                        # render title
    0x224488, P1, "A", P2, "B", P3, "C", P4, "D",         # render points
    0x884422, e12+e01+e02,k,                          # render cross-ratio
    lineWidth=3, grid=True, labels=True, animate=True,
)
```

## Line Cross-Ratio

```python
l = line(0,1,0)

t1 = 0.5
t2 = 0.8
t3 = 1
t4 = 2

li = lambda ti: (e12*ti/2).exp() * l * (-e12*ti/2).exp() 


l1 = li(t1)
l2 = li(t2)
l3 = li(t3)
l4 = li(t4)

```

```python
alg.graph(
    "Drag the points A,B,C,D",                        # render title
    0x224488, l,
    0x114477, l1, l2, l3, l4,
    lineWidth=3, grid=True, labels=True, animate=True,
)
```
