```python
from kingdon import Algebra
import math
import time
```

```python
crossRatio = lambda a,b,c,d: ( (a & c) | ( b & d) ) / ( (b & c) | ( a & d) )
```

You can increase the dimension

```python
d = 2
alg = Algebra(d, 0, 1) 
```

```python
A = alg.vector([1.0, -1.0, *([0.0] * (d - 1))]).dual()
C = alg.vector([1.0, 1.0, *([0.0] * (d - 1))]).dual()
B = lambda: 1.3 * A + C
D = lambda: A + (B-A) * -10.0*math.sin(0.5 * time.perf_counter())
```

```python
O = 0
if d == 2:
    O = alg.vector([1,0, *([1.0] * (d - 1))]).dual()

if d == 3:
    O = alg.vector([1, 0.0, 0.0, -2.0]).dual() & alg.vector([0, 0.0, 1.0, 0.0]).dual()
```

```python
lOA = lambda: O & A
lOB = lambda: O & B
lOC = lambda: O & C
lOD = lambda: O & D
```

```python
k = lambda: f"cross-ratio : {crossRatio(A(),B(),C(),D()):.5f}"
kl = lambda: f"cross-ratio line : {crossRatio(lOA().dual(), lOB().dual(), lOC().dual(), lOD().dual()):.5f}"
```

```python
l = lambda: A() & B()
```

```python
def graph_func():
    return [
        "n-dimension cross-ratio", k,
        "n-dimension cross-ratio", kl,
        O, A, "A", B, "B", C, "C", D, "D",
        0x002211, lOA, lOB, lOC, lOD,
        0x007799, l,
    ]

alg.graph(
    graph_func,
    animate=True,
    scale=0.5,
    lineWidth=6,
    grid=True,
)
```
