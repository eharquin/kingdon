```python
from kingdon import Algebra
import math
import time

#alg = Algebra(1, 0, 1)
alg = Algebra(2, 0, 1) # we use 2d pga for visualization
locals().update(alg.blades)
```

```python
# Construct points
point = lambda x: (e0 + x * e1).dual()
# here we use the dual to projet on e2, inner products will give scalars
crossRatio = lambda a,b,c,d: ( (a & c) * ( b & d) ) / ( (b & c) * ( a & d) ) 
```

```python
AB = lambda: A & B
```

```python
A = point(-1)
B = point(1)
C = point(2)
D = point(3)
```

```python
k = lambda: f"cross-ratio : {crossRatio(A,B,C,D):.3f}"
```

```python
alg.graph(
    "Drag the points A,B,C,D",                         # render titles
    0x00AA88, AB,                                      # render line
    0x224488, A, "A", B, "B", C, "C", D, "D",          # render points
    0x884422, e12+e01+e02,k,                           # render cross-ratio
    lineWidth=3, grid=True, labels=True, animate=True,
)
```
