# Construction 2

```python
Ap = point( 1, -1)
Bp = point( 1,  1)
Cp = point( 3,  1)
Dp = point( 3, -1)
```

```python
V1p = lambda: (Ap&Bp) ^ (Cp&Dp)
V2p = lambda: (Bp&Cp) ^ (Dp&Ap)
```

```python
K1 = lambda: -k1() * (Bp & V1p) / (Ap & V1p)
K2 = lambda: -k2() * (Cp & V2p) / (Bp & V2p)
```

```python
Qp = lambda: K1 * Ap + Bp
Rp = lambda: K2 * Bp + Cp

Pp = lambda: (Qp() & V2p()) ^ (Rp() & V1p()) 
```

```python
ABp = lambda: Ap & V1p
BCp = lambda: Bp & Cp
CDp = lambda: Cp & Dp
DAp = lambda: Dp & Ap
PRp = lambda: Pp() & Rp
PQp = lambda: Pp() & Qp
```
