```python
from kingdon import Algebra
import math
import time

alg = Algebra(3, 0, 1)
locals().update(alg.blades)
```

```python
# Construct points and lines from coefficients
point = lambda x, y, z: (e0 + x*e1 + y*e2 + z*e3).dual()
crossRatio = lambda a,b,c,d: ( (a ^ c) | (b ^ d) ) / ( (b ^ c) | (a ^ d) )
#crossRatio = lambda a,b,c,d: ( (a & c) * ( b & d) ) / ( (b & c) * ( a & d) ) 
```

# Construction 1

```python
A = point( 0, 0.1,  0)
B = point( 1, 0,  0)
C = point( 0, 1,  0)
D = point( 0, 0,  1)
E = point( 1, 0 , 1) 
P = point( 0.5,  0.5, 0.5)
```

```python
ABC = lambda: A & B & C
ABD = lambda: A & B & D
ABE = lambda: A & B & E
ABP = lambda: A & B & P

BCA = lambda: B & C & A
BCD = lambda: B & C & D
BCE = lambda: B & C & E
BCP = lambda: B & C & P

CDA = lambda: C & D & A
CDB = lambda: C & D & B
CDE = lambda: C & D & E
CDP = lambda: C & D & P
```

```python
k1 = lambda: crossRatio(ABC(),ABD(),ABE(),ABP())
k2 = lambda: crossRatio(BCA(),BCD(),BCE(),BCP())
k3 = lambda: crossRatio(CDA(),CDB(),CDE(),CDP())
```

```python
Pr = ABP() ^ BCP() ^ CDP()
```

```python
Pr / 0.225
```

# Construction 2

```python
Ap = point( 0, 0.1,  0)
Bp = point( 1, 0,  0)
Cp = point( 0, 1,  0)
Dp = point( 0, 0,  1)
Ep = point( 1, 0 , 1) 
```

```python
ABCp = lambda: Ap & Bp & Cp
ABDp = lambda: Ap & Bp & Dp
ABEp = lambda: Ap & Bp & Ep

BCAp = lambda: Bp & Cp & Ap
BCDp = lambda: Bp & Cp & Dp
BCEp = lambda: Bp & Cp & Ep

CDAp = lambda: Cp & Dp & Ap
CDBp = lambda: Cp & Dp & Bp
CDEp = lambda: Cp & Dp & Ep
```

```python
K1 = lambda: (-k1() * ( (ABDp() ^ ABEp()) / (ABCp() ^ ABEp()) ) ).e
K2 = lambda: (-k2() * ( (BCDp() ^ BCEp()) / (BCAp() ^ BCEp()) ) ).e
K3 = lambda: (-k3() * ( (CDBp() ^ CDEp()) / (CDAp() ^ CDEp()) ) ).e
```

```python
K1()
```

```python
K2()
```

```python
K3()
```

```python
ABPp = lambda: K1() * ABCp() + ABDp()
BCPp = lambda: K2() * BCAp() + BCDp()
CDPp = lambda: K3() * CDAp() + CDBp()

Pp = lambda: ABPp() ^ BCPp() ^ CDPp()
```
