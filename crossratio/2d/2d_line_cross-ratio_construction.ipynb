```python
from kingdon import Algebra
import math
import time

alg = Algebra(2, 0, 1)
locals().update(alg.blades)
```

```python
# Construct points and lines from coefficients
point = lambda x, y: (e0 + x*e1 + y*e2).dual()
crossRatio = lambda a,b,c,d: ( (a ^ c) | (b ^ d) ) / ( (b ^ c) | (a ^ d) )
```

# Construction 1

```python
A = point(-3, -1)
B = point(-3,  1)
C = point( -1,  1)
D = point( -1, -1)
P = point( -2.2,  0.5)
```

```python
BA = lambda: B & A
CA = lambda: C & A
DA = lambda: D & A
PA = lambda: P & A

AB = lambda: A & B
CB = lambda: C & B
DB = lambda: D & B
PB = lambda: P & B
```

```python
k1 = lambda: crossRatio(BA(),CA(),DA(),PA())
k2 = lambda: crossRatio(AB(),CB(),DB(),PB())
```

# Construction 2

```python
Ap = point( 1, -1)
Bp = point( 1,  1)
Cp = point( 3,  1)
Dp = point( 3, -1)
```

```python
BAp = lambda: Bp & Ap
CAp = lambda: Cp & Ap
DAp = lambda: Dp & Ap

ABp = lambda: Ap & Bp
CBp = lambda: Cp & Bp
DBp = lambda: Dp & Bp
```

```python
K1 = lambda: -k1() * (CAp() ^ DAp()) / (BAp() ^ DAp())
K2 = lambda: -k2() * (CBp() ^ DBp()) / (ABp() ^ DBp())
```

```python
lP1 = lambda: K1() * BAp + CAp
lP2 = lambda: K2() * ABp + CBp

Pp = lambda: lP1() ^ lP2()
```

```python
alg.graph(
    "Drag the points A,B,C,D,P",                        # render title

    #0x884422, AB, BC, CD, DA, PR, PQ, V1, "V1", V2, "V2", Q, "Q", R, "R",  # CONSTRUCTION 1
    0x224488, A, "A", B, "B", C, "C", D, "D",         # render points C1
    0x305a7f, [A,B,C,D],
    0x887711, P, "P",

    #0x884422, ABp, BCp, CDp, DAp, PRp, PQp, V1p, "V1", V2p, "V2", Qp, "Q", Rp, "R",  # CONSTRUCTION 1
    0x884422, lP1, lP2,
    0x224488, Ap, "Ap", Bp, "Bp", Cp, "Cp", Dp, "Dp",
    0x305a7f, [Ap,Bp,Cp,Dp],
    0x887711, Pp, "Pp",
    
    lineWidth=3, grid=True, labels=True, animate=True,
)
```
