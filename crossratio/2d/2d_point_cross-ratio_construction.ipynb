```python
from kingdon import Algebra
import math
import time

alg = Algebra(2, 0, 1)
locals().update(alg.blades)
```

```python
# Construct points and lines from coefficients
point = lambda x, y: (e0 + x*e1 + y*e2).dual()
crossRatio = lambda a,b,c,d: ( (a & c) | ( b & d) ) / ( (b & c) | ( a & d) ) 
```

# Construction 1

```python
A = point(-3, -1)
B = point(-3,  1)
C = point( -1,  1)
D = point( -1, -1)
P = point( -2.2,  0.5)
```

```python
V1 = lambda: (A&B) ^ (C&D)
V2 = lambda: (B&C) ^ (D&A)
```

```python
Q = lambda: (P&V2) ^ (A&B)
R = lambda: (P&V1) ^ (B&C)
```

```python
k1 = lambda: crossRatio(A,B,V1,Q)
k2 = lambda: crossRatio(B,C,V2,R)
```

```python
AB = lambda: A & V1
BC = lambda: B & C
CD = lambda: C & D
DA = lambda: D & A
PR = lambda: P & R
PQ = lambda: P & Q
```

# Construction 2

```python
Ap = point( 1, -1)
Bp = point( 1,  1)
Cp = point( 3,  1)
Dp = point( 3, -1)
```

```python
V1p = lambda: (Ap&Bp) ^ (Cp&Dp)
V2p = lambda: (Bp&Cp) ^ (Dp&Ap)
```

```python
K1 = lambda: -k1() * (Bp & V1p) / (Ap & V1p)
K2 = lambda: -k2() * (Cp & V2p) / (Bp & V2p)
```

```python
Qp = lambda: K1 * Ap + Bp
Rp = lambda: K2 * Bp + Cp

Pp = lambda: (Qp() & V2p()) ^ (Rp() & V1p()) 
```

```python
ABp = lambda: Ap & V1p
BCp = lambda: Bp & Cp
CDp = lambda: Cp & Dp
DAp = lambda: Dp & Ap
PRp = lambda: Pp() & Rp
PQp = lambda: Pp() & Qp
```
