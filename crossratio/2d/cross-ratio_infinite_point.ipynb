```python
from kingdon import Algebra
import math
import time

alg = Algebra(2, 0, 1)
locals().update(alg.blades)
```

```python
# Construct points and lines from coefficients
point = lambda x, y: (e0 + x*e1 + y*e2).dual()
crossRatio = lambda a,b,c,d: ( (a & c) | ( b & d) ) / ( (b & c) | ( a & d) ) 
```

```python
A = point(-1, 0)
B = point(1, 0.8)
D = lambda: B - A # direction of the line same as (B-A) for the cr formula
C = lambda:  A + 1.7 * D()
```

```python
AB = lambda: A & B
```

```python
k = lambda: f"cross-ratio : {crossRatio(A,B,C,D).grade(0):.3f}" # Here we use .grade(0) to clear near zero coefs
```

```python
pointOnLine = lambda l, pt : pt.proj(l) 
```

```python
alg.graph(
    "Drag the points A,B,C,D",                        # render title
    0x00AA88, AB,                                     # render line
    0x224488, A, "A", B, "B", C, "C",                 # render points
    0x884422, e12+e01+e02,k,                          # render cross-ratio
    0x882244, D, "D",                                 # render vectors
    lineWidth=3, grid=True, labels=True, animate=True,
)
```
